### Compare June and December data across all years 

```python
from scipy import stats
```

```python
# Filter data for desired months # Filter by month in index directly
June = df[df.index.month == 6]
June
```

```python
Dec = df[df.index.month == 12]
Dec
```

```python
# Identify the average temperature for June # 74.944118
June.describe()
```

```python
# Identify the average temperature for December # 71.041529
Dec.describe()
```

```python
# Create collections of temperature data
June_temp = June['tobs'].to_list()
Dec_temp = Dec['tobs'].to_list()
```

```python
# Since the lengths of the tobs list are different between June and Dec so can not run paired ttest. I am running unpaired indepedent t-test below. 
from scipy import stats
stats.ttest_ind(June_temp, Dec_temp)
```

### Analysis

# In years 2010-2017, the mean temp in June and Dec differs by 3.9F. Then, an unpaired t-test was conducted. With an extremely low p-value, the difference is deemed statistically significant.
